## File Handling in Python
- File handling is the process of creating, reading, writing, and managing files in a program
- Makes system robust, secure, maintainable

```python
# Open, read & write 

# We have to open a file before reading and writing 
# f = open('file','mode')

file1 = open('simple.txt')   # by default mode read

# RECOMMENDED : f = open('path/name.txt','mode') 
file2 = open('employment Data.csv', 'r')


# Perform operation 
data = file1.read()
print(data,type(data))  # returns string stream 


# Step 3 : imp > we have to close the file that we open 
file1.close()
file2.close()
```
