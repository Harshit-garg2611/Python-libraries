# Matplotlib : 

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(['a',2,3,4],'ro')
plt.ylabel("Sales")
plt.xlabel("Year")
plt.show()

# If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you.
```

```python
# to plot x versus y, you can write:

plt.plot([1900,1920,1940,1960],[10,20,30,40])
plt.show()
```

```python
x = [1900,1920,1940,1960,1970,1980]
hr = [15,18,25,50,125,75]
dl = [20,44,65,17,34,98]


# plot multiple in same graph
# plt.plot(x,hr, label='Haryana',color='g', linewidth=2, linestyle='--')
plt.plot(x,dl,label='Delhi', color='orange', marker='*') # we can use short hand notation as color,marker,line style as :

plt.plot(x,hr, 'g*--',label='Haryana', linewidth=2)

 
# Title :
plt.title("Sales for different states",
          fontdict={'fontname':'comic sans ms', 'color':'Red','fontsize':20})

# labels
plt.ylabel("Sales", fontdict={'fontname':'comic sans ms', 'color':'Blue','fontsize':20})
plt.xlabel("Year", fontdict={'fontname':'comic sans ms', 'color':'Blue','fontsize':20})

# To manage range distribution in axis 
plt.xticks([1900,1920,1940,1960,1980,2000,2020])
plt.yticks([x*10 for x in range(1,16)])




plt.legend(loc ='best') 
# best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 
# 'lower center', 'upper center', 'center' 



# plt.savefig('sales.png',transparent=1, dpi=1000)
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0, 1, 0.1)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# inorder to create different figures , we use plt.figure()

x = [1900,1920,1940,1960,1970,1980]
hr = [15,18,25,50,125,75]
dl = [20,44,65,17,34,98]



plt.figure()
plt.plot(x,dl,label='Delhi', color='orange', marker='*') 



# start new figure 
plt.figure()
plt.plot(x,hr, 'g*--',label='Haryana', linewidth=2)

 
# Title :
plt.title("Sales for different states",
          fontdict={'fontname':'comic sans ms', 'color':'Red','fontsize':20})

# labels
plt.ylabel("Sales", fontdict={'fontname':'comic sans ms', 'color':'Blue','fontsize':20})
plt.xlabel("Year", fontdict={'fontname':'comic sans ms', 'color':'Blue','fontsize':20})

# To manage range distribution in axis 
plt.xticks([1900,1920,1940,1960,1980,2000,2020])
plt.yticks([x*10 for x in range(1,16)])




plt.legend(loc ='best') 
# best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 
# 'lower center', 'upper center', 'center' 



# plt.savefig('sales.png',transparent=1, dpi=1000)
plt.show()
```

```python
# inorder to create subplots :

import numpy as np
t = np.arange(0, 1, 0.1)

plt.figure()

plt.subplot(1,3,1)  # arg : row, col, index
plt.plot(t, t, 'r--')



plt.subplot(1,3,2)
plt.plot(t, t**2, 'bs')



plt.subplot(1,3,3)
plt.plot(t, t**3, 'g^')


plt.show()
```
